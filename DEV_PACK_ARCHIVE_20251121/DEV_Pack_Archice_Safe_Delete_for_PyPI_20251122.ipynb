```python
import pandas
import PY3_Pandas2_Read_Directory
```

```python
OBJ_PY3_Pandas2_Read_Directory = PY3_Pandas2_Read_Directory.PY3_Pandas2_Read_Directory()
```

```python
def Read_for_Retain_PyPI_Packages(Directory_for_Retain:str):

    LIST_DataFrame:list = []
    # ..............................................
    JSON_Records = OBJ_PY3_Pandas2_Read_Directory.Read_Directory(Directory_for_Retain, Header = None)
    # ..............................................0
    for Record in JSON_Records:
        for df in Record["dataframe"].values():
            LIST_DataFrame.append(df)
    # ..............................................
    return pandas.concat(LIST_DataFrame, ignore_index = True)

def Read_for_Delete_PyPI_Packages(Directory_for_Delete:str):

    Pandas_DataFrame = Read_for_Retain_PyPI_Packages(Directory_for_Retain = Directory_for_Delete)
    # ..............................................
    return Pandas_DataFrame
```

```python
Retain_PyPI_Packages_DataFrame = Read_for_Retain_PyPI_Packages("./Pack_Archive_Safe_Delete_for_Retain")
Retain_PyPI_Packages_DataFrame[0] = Retain_PyPI_Packages_DataFrame[0].apply(lambda x: x.replace('_', '-'))
Retain_PyPI_Packages_DataFrame
```

```python
Delete_PyPI_Packages_DataFrame = Read_for_Delete_PyPI_Packages("./Pack_Archive_Safe_Delete_for_Delete")
Delete_PyPI_Packages_DataFrame[0] = Delete_PyPI_Packages_DataFrame[0].apply(lambda x: x.replace('_', '-'))
Delete_PyPI_Packages_DataFrame
```

```python
Confirm_Delete:list = []
# ..................................................
for idx, row in Delete_PyPI_Packages_DataFrame.iterrows():
    PyPI_Package = row[0]
    if (PyPI_Package not in Retain_PyPI_Packages_DataFrame[0].values):
        Confirm_Delete.append(PyPI_Package)
# ..................................................
Confirm_Delete = list(set(Confirm_Delete))
Confirm_Delete.sort()
Confirm_Delete
```
