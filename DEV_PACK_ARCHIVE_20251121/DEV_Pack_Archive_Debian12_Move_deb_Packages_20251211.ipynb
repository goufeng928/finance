```python
Packages_Path:str = "./Packages.gz"
Packages_Filename:list = Parse_Mirrors_Packages.Find_Filename(Path = Packages_Path)
# ..................................................
print("[\n ......")
a:int = 0
b:int = len(Packages_Filename[10:15])  # 列表 (List) 切片
c:int = b - 1
while (a <= c):
    Element = Packages_Filename[a]
    Element = str(Element)
    print(" \"%s\"," % str(Element))
    a = a + 1
print(" ......\n]")
```

```python
DataFrame_List:list = [pandas.DataFrame(), pandas.DataFrame()]
```

```python
df = pandas.DataFrame({"filename": Packages_Filename})
df["filename" ] = df["filename"].str.split(':')                                     # "Filename: pool/main/0/0ad/0ad_0.0.26-3_amd64.deb"      ---> ["Filename", " pool/main/0/0ad/0ad_0.0.26-3_amd64.deb"]
df["filename" ] = df["filename"].apply(lambda x: x[1] if len(x) >= 2 else x[0])     # ["Filename", " pool/main/0/0ad/0ad_0.0.26-3_amd64.deb"] ---> " pool/main/0/0ad/0ad_0.0.26-3_amd64.deb"
df["filename" ] = df["filename"].str.split('_')                                     # " pool/main/0/0ad/0ad_0.0.26-3_amd64.deb"               ---> [" pool/main/0/0ad/0ad_0.0.26-3", "amd64.deb"]
df["filename" ] = df["filename"].apply(lambda x: x[0])                              # [" pool/main/0/0ad/0ad_0.0.26-3", "amd64.deb"]          ---> " pool/main/0/0ad/0ad_0.0.26-3"
df["filename" ] = df["filename"].apply(lambda x: x[1:] if x[0] == str(' ') else x)  # " pool/main/0/0ad/0ad_0.0.26-3"                         ---> "pool/main/0/0ad/0ad_0.0.26-3"
df["directory"] = df["filename"].apply(lambda x: os.path.dirname(x))
df["basename" ] = df["filename"].apply(lambda x: x.split('/')[-1])
# ..................................................
DataFrame_List[0] = df
DataFrame_List[0]
```

```python
Entry:str          = "./Pack_Archive_Debian12_Move_deb_Packages"
Filename_List:list = os.listdir(Entry)
# ..................................................
df = pandas.DataFrame({"filename": Filename_List})
df = df[df["filename"].str.contains("keep.the.folder") != True]
df["basename"] = df["filename"].str.split('_')         # "0ad_0.0.26-3_amd64.deb"         ---> ["0ad", "0.0.26-3", "amd64.deb"]
df["basename"] = df["basename"].apply(lambda x: x[0])  # ["0ad", "0.0.26-3", "amd64.deb"] ---> "0ad"
# ..................................................
DataFrame_List[1] = df
DataFrame_List[1]
```

```python
if ("filename" in DataFrame_List[0].columns):
    DataFrame_List[0] = DataFrame_List[0].drop("filename", axis = 1)
# ..................................................
df = pandas.merge(left = DataFrame_List[1], right = DataFrame_List[0], how="left", on = "basename")
df = df[df["directory"].isna() != True]
df
```

```python
df["directory"] = df["directory"].fillna(value = str(''))
df["directory"].apply(lambda x: os.makedirs("./dist/%s" % x, exist_ok = True))
```

```python
df.apply(lambda x: os.rename(f"""./{Entry}/{x["filename"]}""", f"""./dist/{x["directory"]}/{x["filename"]}"""), axis = 1)
```
