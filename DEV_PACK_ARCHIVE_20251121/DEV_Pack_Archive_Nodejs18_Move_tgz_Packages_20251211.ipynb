```python
import os
import sys
# ..................................................
import pandas
# ..................................................
import PY3_Parse_Nodejs18_Package_Lock_Packages
```

```python
WORK_PATH:str         = "Pack_Archive_Nodejs18_Move_tgz_Packages"
TGZ_DIR_PATH:str      = "Pack_Archive_Nodejs18_Move_tgz_Packages/tgz"
PACKAGE_LOCK_PATH:str = "Pack_Archive_Nodejs18_Move_tgz_Packages/package-lock.json"
```

```python
Nodejs18_Package_Lock = PY3_Parse_Nodejs18_Package_Lock_Packages.PY3_Parse_Nodejs18_Package_Lock_Packages()
```

```python
Tgz_for_URL:dict   = {"url": Nodejs18_Package_Lock.Find_All_Tgz(PACKAGE_LOCK_PATH)}
Tgz_for_Local:dict = {"tgz": os.listdir(TGZ_DIR_PATH)}
```

```python
df1 = pandas.DataFrame(Tgz_for_URL)
df1 = df1.drop_duplicates()
df1
```

```python
df2 = pandas.DataFrame(Tgz_for_Local)
df2 = df2.drop_duplicates()
df2
```

```python
df1["tgz"] = df1["url"].apply(lambda x: x.replace("https://", ''))
df1["tgz"] = df1["tgz"].apply(lambda x: x.replace("http://", ''))
df1["tgz"] = df1["tgz"].apply(lambda x: os.path.basename(x))  # "/path/to/example.tgz" => "example.tgz"
# ..................................................
df1["path"] = df1["url"].apply(lambda x: x.replace("https://", ''))
df1["path"] = df1["path"].apply(lambda x: x.replace("http://", ''))
df1["path"] = df1["path"].apply(lambda x: os.path.dirname(x))  # "/path/to/example.tgz" => "/path/to"
df1
```

```python
df0 = pandas.merge(left = df2, right = df1, how = "left", on = "tgz")
df0
```

```python
Total:int = df0.shape[0]
Count:int = 1
for idx, row in df0.iterrows():
    if (Count % 10 == 0):
        sys.stdout.write(f"\r[DEBUG] PROCESSING: {Count}/{Total}")
        sys.stdout.flush()
    # ..............................................
    Path:str = row["path"]
    Tgz:str  = row["tgz" ]
    # ..............................................
    os.makedirs(f"{WORK_PATH}/{Path}", exist_ok = True)
    os.rename(f"{TGZ_DIR_PATH}/{Tgz}", f"{WORK_PATH}/{Path}/{Tgz}")
    # ..............................................
    Count = Count + 1
```
