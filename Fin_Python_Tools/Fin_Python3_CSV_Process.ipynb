```python
import pandas as pd
```

```python
CsvFile = "./model_rising_horizontal_sh_000001_20141229_20150206.csv"
```

```python
CsvProject = pd.read_csv(CsvFile, header=1)
```

```python
df = pd.DataFrame(CsvProject)
```

```python
df.dtypes
```

```python
df
```

```python
df["名称"] = "上证指数"
df["代码"] = "000001"
```

```python
df
```

```python
# 组织 CSV 行内容。

for df_date, \
    df_name, \
    df_code, \
    df_open, \
    df_high, \
    df_low, \
    df_close, \
    df_rise_fall_amt, \
    df_rise_fall_rate, \
    df_volume, \
    df_trading_amount in \
    zip(df["日期"], \
        df["名称"], \
        df["代码"], \
        df["开盘"], \
        df["最高"], \
        df["最低"], \
        df["收盘"], \
        df["涨跌额"], \
        df["涨跌幅"], \
        df["成交量(手)"], \
        df["成交金额(万)"]):
    
    CsvDate = df_date
    CsvName = df_name
    CsvCode = df_code
    CsvOpen = df_open
    CsvHigh = df_high
    CsvLow = df_low
    CsvClose = df_close
    CsvRiseFallAmt = df_rise_fall_amt
    CsvRiseFallRate = df_rise_fall_rate
    CsvVolume = df_volume
    CsvTradingAmount = df_trading_amount

    CsvRow = "%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s" % \
             (CsvDate, CsvName, CsvCode, CsvOpen, CsvHigh, CsvLow, CsvClose, \
              CsvRiseFallAmt, CsvRiseFallRate, CsvVolume, CsvTradingAmount)

    print(CsvRow)
```
