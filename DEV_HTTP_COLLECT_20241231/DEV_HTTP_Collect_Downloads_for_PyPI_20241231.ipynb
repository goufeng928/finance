```python
import json
import os
import sys
# ..................................................
import pandas
# ..................................................
import PY3_Requests2_BS4_HTTP_Collect
import PY3_HTTP_URL_Path_Operate
```

```python
Requests2_BS4_HTTP_Collect = PY3_Requests2_BS4_HTTP_Collect.PY3_Requests2_BS4_HTTP_Collect()
HTTP_URL_Path_Operate      = PY3_HTTP_URL_Path_Operate.PY3_HTTP_URL_Path_Operate()
```

```python
FILE_Json = open("HTTP_Headers.json", mode = 'r', encoding = "utf-8")
TEXT_Json = FILE_Json.read()
FILE_Json.close()
LIST_Json = json.loads(TEXT_Json)
# ..................................................
HTTP_Headers = LIST_Json["mirrors.tuna.tsinghua.edu.cn/pypi/web/simple"][0]
```

```python
Target_DataFrame = pandas.read_csv("./Collect_Downloads_for_PyPI_Targets.csv", dtype = str)
Target_DataFrame
```

```python
LIST_Collected_All:list = []
# ..................................................
Total:int = Target_DataFrame["name"].count()
Count:int = 1
# ..................................................
Requests2_BS4_HTTP_Collect.HTTP_Headers = HTTP_Headers
# ..................................................
for idx, row in Target_DataFrame.iterrows():
    Package_Name:str    = row["name"]
    Package_Version:str = row["version"]
    # ..............................................
    sys.stdout.write(f"\r[DEBUG] Processing -> Package Name: {Package_Name}, Package Version: {Package_Version} ({Count}/{Total})")
    sys.stdout.flush()
    # ..............................................
    URL = "https://mirrors.tuna.tsinghua.edu.cn/pypi/web/simple" + '/' + Package_Name
    # ..............................................
    LIST_Collected = Requests2_BS4_HTTP_Collect.GET_Tags_a_Attrs_href(URL = URL)      # 获取 "标签 a" 的 "属性 href" 的值
    LIST_Collected = [href for href in LIST_Collected if (Package_Version in href)]  # 筛选 Package 版本
    # ..............................................
    LIST_Collected_All.extend(LIST_Collected)
    # ..............................................
    Count = Count + 1
```

```python
Collected_DataFrame = pandas.DataFrame({"url": LIST_Collected_All})
Collected_DataFrame
```

```python
Collected_DataFrame["basename"] = Collected_DataFrame["url"     ].apply(lambda x: HTTP_URL_Path_Operate.Get_URL_Basename(x))
Collected_DataFrame["basename"] = Collected_DataFrame["basename"].apply(lambda x: x.split('#')[0] if ('#' in x) else x)
Collected_DataFrame
```

```python
Total:int = Collected_DataFrame["url"].count()
Count:int = 1
# ..................................................
Requests2_BS4_HTTP_Collect.HTTP_Headers = HTTP_Headers
# ..................................................
for idx, row in Collected_DataFrame.iterrows():
    URL:str      = row["url"     ]
    Basename:str = row["basename"]
    # ..............................................
    sys.stdout.write(f"\r[DEBUG] Processing -> Basename: {Basename} ({Count}/{Total})")
    sys.stdout.flush()
    if (os.path.exists(f"./Collect_Downloads_for_PyPI/{Basename}") != True):
        Requests2_BS4_HTTP_Collect.GET_Download_File(URL = URL, Save_Path = f"./Collect_Downloads_for_PyPI/{Basename}")
    # ..............................................
    Count = Count + 1
```
